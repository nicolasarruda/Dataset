```python
%matplotlib inline
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import numpy as np
plt.style.use('ggplot')
```

```python
renting = pd.read_csv('2020-06-03-SP-renting-RUIDO.csv', sep = ";")
```

```python
renting
```

```python
renting.corr().style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'),axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
KMModel = kmeans.fit(renting)
```

```python
clusters = kmeans.cluster_centers_
```

```python
KMModel.labels_
```

```python
pdrenting = pd.crosstab(renting.id,KMModel.labels_)
```
